

*   텐서플로 기반의 딥러닝 아키텍처는 **은닉층의 활성화 함수**로 시그모이드가 아니, **relu 함수**를 주로 사용하며, **출력층**에서도 시그모이드 함수 값을 그대로 사용하지 않고, 출력 값의 확률 분포를 나타내는 **소프트맥스 softmax** 가 주로 사용됨.

*   **relu 함수** : 입력 값이 0 이하면, 0을 내보내고, 0보다 크면 입력 값을 그대로 출력. 학습 속도가 빠르다는 장점

* **소프트맥스 softmax 함수** : 분모 부분에서 모든 입력 값을 더하고, 분자에서는 특정한 입력 값 하나만 들어감. 특징은,

    1.   입력 값의 대소 순서와 출력 값의 대소 순서가 같음
    2.   각각의 입력 값에 대한 출력은 0과 1 사이의 값이 나옴
    3.   각각의 입력에 대한 출력 값들을 모두 더하면 1. 이는 각각의 출력 값이 확률 Probability을 나타냄

# tensorflow_v1_2.py 파일에 작성된 코드 내용

```python
# import tensorflow as tf

import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

import numpy as np

loaded_data = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/AI/ML_DL_NeoWizard_Youtube/lecture_file/data-01.csv', delimiter=',')

x_data = loaded_data[ :, 0:-1]
t_data = loaded_data[ :, [-1]]

print("x_data.shape = ", x_data.shape)
print("t_data.shape = ", t_data.shape)


W = tf.Variable(tf.random_normal([3, 1]))
b = tf.Variable(tf.random_normal([1]))

X = tf.placeholder(tf.float32, [None, 3])
T = tf.placeholder(tf.float32, [None, 1])

y = tf.matmul(X, W) + b

loss = tf.reduce_mean(tf.square(y - T))

learning_rate = 1e-5

optimizer = tf.train.GradientDescentOptimizer(learning_rate)

train = optimizer.minimize(loss)


with tf.Session() as sess :

  sess.run(tf.global_vairiables_initializer())

  for step in range(8001) :

    loss_val, y_val, _ = sess.run([loss, y, train], feed_dict={X: x_data, T : t_data})

    if step % 400 == 0 :
      print("step = ", step, ", loss_val = ", loss_val)

  print("\nPrediction is ", sess.run(y, feed_dict={X : [ [100, 98, 81] ]}))
```
